### 1. Introduccin a Avogadro 

Avogadro es un **editor molecular** para construir y ver molculas y materiales en 3D.

#### Instalcin 

Para evitar conflictos con versiones de OpenBabel ya instaladas en nuestra computadora, es conveniente hacer la instalacin en un nuevo entorno de conda

* Descargar e instalar anaconda: https://www.anaconda.com/products/individual

* Crear un nuevo entorno de conda:  
``` bash
> conda create -n myenv
```

* Activar el entorno: 
``` bash
> conda activate myenv 
```

* Instalar avogadro dentro del entorno: 
``` bash
> conda install -c rpmuller avogadro
```


#### Una vez instalado.....

Cuando abre inicialmente Avogadro, se le presentar una pantalla como la que se muestra a continuacin:




#### Abrir un archivo 

Lo primero que probablemente querr hacer es abrir un archivo y visualizar una molcula. Para ello haga clic en el men **File** y seleccione **Open...**



Se admite una gran cantidad de tipos de archivos, incluidos CML, XYZ, SDF, Mol2, PDB, etc. Se suministran varias molculas de ejemplo con Avogadro.




La captura de pantalla anterior muestra el archivo etanol.cml usando el tipo de visualizacin predeterminado "Ball and Stick". Tenga en cuenta que cuando se abre un archivo nuevo, Avogadro cambia de modo "Drawing" a modo "Navigation", lo que le permite ver la molcula sin editarla.

#### Navagacin

Puede acercar/alejar la vista con la rueda de desplazamiento del mouse o manteniendo presionado el botn central del mouse y moviendo el cursor del mouse arriba/abajo. Puede rotar la vista manteniendo presionado el botn izquierdo  y moviendo el cursor del mouse. Tambin puede transladar la vista manteniendo presionado el botn derecho y moviendo el cursor del mouse.

#### Creando moleculas

Las molculas se pueden construir y editar con la herramienta de dibujo (Drawing mode).




Puede comenzar a crear una molcula haciendo clic izquierdo en la pantalla negra. Esto generar un tomo de carbono. Haciendo clic derecho sobre el tomo lo borrar.





Al hacer clic con el botn izquierdo y arrastrar el mouse, se generar un enlace con otro tomo de carbono:




Se pueden seleccionar diferentes elementos en el men desplegable "Elements":





Por ejemplo, si desea crear una molcula de agua, puede escribir "O" o seleccionar "Oxgeno (8)" en el men desplegable y hacer clic izquierdo en la pantalla negra.





Si la casilla "Ajust hidrogens" est marcada, los hidrgenos en la molcula se ajustan automticamente para satisfacer la valencia (como se muestra arriba).





El orden de enlace se puede cambiar a travs del men desplegable "Bond order" o escribiendo los nmeros "1", "2" o "3". Los enlaces tambin se pueden agregar a una molcula haciendo clic con el botn izquierdo en un enlace que ya se ha creado. Al hacer clic con el botn derecho en un enlace, se eliminar el enlace y el tomo al que est unido.



Por ejemplo, para crear dixido de carbono, se debe dibujar la estructura general ("O-C-O") y luego, al hacer clic con el botn izquierdo en los enlaces de la molcula, se crearn enlaces dobles.





En el men "Extensions", la geometra de una molcula se puede optimizar seleccionando "Optimize geometry".



Luego de finalizada la optimizacin, la molcula de dioxido de carbono adquiere una conformacin lineal. 
Avogadro viene equipado con diferentes campos de fuerza o Force Fields (ya estudiaremos mas adelante los campos de fuerza). Por defecto Avogadro utiliza el campo de fuerza de mecnica molecular UFF (Universal Force Field) para realizar las optimizaciones de geometra. Este campo de fuerza puede optimizar la geometra de todos los elementos y funciona bien con materiales inorgnicos y materiales organometlicos.


Para mas informacin sobre las funcionalidades de Avogadro puede consultar la pgina del programa: https://avogadro.cc/

### 2. Introduccin a PyMOL

#### Instalacin
Puede descargar el instalador de Pymol apropiado para su sistema operativo de: https://pymol.org/2/#download

#### La interfaz PyMOL
Cuando se abre PyMOL, aparecen dos ventanas. La ventana ms pequea (llamada "GUI externa") contiene la barra de men (Archivo, Editar, Ayuda, Mostrar, etc.), botones de acceso directo para comandos comunes y la lnea de comando.<br>
La segunda ventana es el **Visualizador de PyMOL** donde se muestran los modelos 3D y el usuario interacta (por ejemplo, gira) y manipula el modelo.Los objetos que renderiza PyMOL en 3D se cargan desde archivos de coordenadas que describen las posiciones de los tomos individuales en la molcula.<br> 



PyMOL puede mostrar ms de un objeto a la vez, el **Panel de control de objetos** permite ajustar los modos de visualizacin, los colores, las etiquetas, etc para cada objeto. Despus de cada nombre de objeto hay un conjunto de botones de comando que controlan el objeto. Estos son los botones y algunas de sus opciones:

**A** - Acciones: cambiar el nombre, duplicar, eliminar, aplicar ajustes preestablecidos, realizar clculos <br>
**S** - Mostrar: cambia la representacin de los tomos, por ejemplo,  **wireframe** o **stick** para visualizar todos los tomos y enlaces de la proteina o **cartoon** para visualizar los elementos de estructura secundaria <br>
**H** - Ocultar: las cosas que se muestran usando S se acumulan y no reemplazan automticamente la ltima vista. H es lo opuesto a S y oculta representaciones no deseadas <br>
**L** - Etiqueta: Etiqueta tomos, residuos, etc <br>
**C** - Color: cambia el color de los tomos y grupos <br>

La esquina inferior derecha del Visor contiene una gua para usar el mouse, as como una poderosa herramienta de seleccin. Tambin hay otra lnea de comando en la parte inferior del Visor (PyMOL>).<br>
