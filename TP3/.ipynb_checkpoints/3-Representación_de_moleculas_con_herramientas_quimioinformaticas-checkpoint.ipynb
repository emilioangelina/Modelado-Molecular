### Que es la quimioinformtica ?

Es el conjunto de recursos informticos para transformar DATOS en INFORMACION e informacin en CONOCIMIENTO, con el propsito de tomar ms rpidas y mejores decisiones en el mbito de la identificacin y optimizacin de frmacos. 



Un aspecto importante en quimioinformtica es como representamos las molculas de manera de poder acceder rpidamente a la informacin contenida en ellas, mediante ordenadores. 
Una representacin molecular reduce la dimensionalidad de una estructura molecular reteniendo al mismo tiempo la informacin qumica importante. Por ejemplo, una frmula qumica (ej. $C_6$$H_6$) reduce una molcula tridimensional a una cadena de caracteres, pero es una entrada ambigua para un algoritmo quimioinformtico. Es decir que la estructura atmica y molecular debe poder convertirse a un formato legible por mquina. 

### Representacin 1-D, 2-D y 3-D de molculas

En quimioinformtica, la informacin estructural de una molcula se puede dar como una cadena 1D de caracteres (denominada SMILES), un dibujo 2D (topologa molecular) o una representacin 3D. El tipo de representacin depender del tipo de propiedad molecular que se quiera predecir. Las herramientas quimioinformticas permiten convertir rpidamente de una representacin en otra.  



### SMILES

Es la representacin ms sencilla de una molcula. Se escribe la molcula como una cadena de caracteres en una sola lnea de texto, siguiendo una serie de reglas: 
- Los diferentes elementos qumicos se representan mediante sus smbolos qumicos, generalmente se ignora el hidrgeno. 
- Los enlaces simples no se muestran; para enlaces dobles, triples y cudruples se usa '=', '#', '$', respectivamente. 
- Los tomos que estn enlazados se escriben adyacentes.  
- Las estructuras de anillo se escriben rompiendo cada anillo en un punto arbitrario para crear una estructura acclica y agregando etiquetas numricas de cierre de anillo para mostrar la conectividad entre tomos no adyacentes.
- La aromacidad se ilustra comnmente escribiendo los tomos constituyentes B, C, N, O, P y S en minsculas b, c, n, o, p y s, respectivamente.

### MOL y SDF
Es un formato de archivo que guarda informacin sobre los tomos, enlaces, conectividad y coordenadas de una molcula.
Un archivo MOL tpico consta de cierta informacin de ENCABEZADO, seguido de un bloque de informacin de los ATOMOS, luego un bloque para los ENLACES, seguidos de secciones para informacin ms compleja. Las representaciones MOL pueden contener informacin sobre la estructura 2D o 3D de la molcula. Este formato est tan extendido que la mayora de los sistemas de software de quimioinformtica pueden leerlo.

La diferencia con el formato SDF (Structure Data Files) es que ste generalmente contine muchos compuestos en formato MOL puestos en un nico archivo, junto con alguna informacin adicional sobre los compuestos. Generalmente se usan para distribuir bibliotecas de compuestos.

A continuacin, veamos si el Ibuprofeno cumple con las regla de los 5 de Lipinski (RO5). 

<div class="alert alert-block alert-info">
La RO5 es una regla emprica que permite evaluar cualitativamente cmo de adecuado podra resultar un compuesto qumico para cumplir alguna determinada funcin farmacolgica o actividad biolgica una vez que es ingerido como medicamento para consumo <b>oral</b> en seres humanos. Esta regla fue formulada por Christopher A. Lipinski en el ao 1997, basado en la observacin de que la mayora de los compuestos qumicos utilizados en medicamentos son molculas relativamente pequeas y lipoflicas.
La regla describe las propiedades moleculares que tienen importancia farmacocintica en el cuerpo humano, incluyendo la absorcin, distribucin, metabolismo, y excrecin ("ADME"). 
Tal y como lo indica la regla de Lipinski, en general, un principio activo y para que sea posible su administracin por va oral no debe violar ms de una de las siguientes consideraciones:

<ul>
<li>No debe contener ms de 5 donadores de hidrgeno (tomos de nitrgeno u oxgeno con al menos un tomo de hidrgeno).</li>
<li>No debe contener ms de 10 aceptores de hidrgeno (tomos de nitrgeno, oxgeno o flor).</li>
<li>Debe poseer un peso molecular inferior a 500.</li>
<li>Debe poseer un coeficiente de reparto octanol-agua (log P) inferior a 5. </li>
</ul>

Ntese que todos los nmeros mencionados en estas condiciones son mltiplos de cinco, lo cual es el origen del nombre de esta regla.

</div>

Importamos el modulo rdkit.Chem.Descriptors que permite calcular otras propiedades moleculares: 

```python
from rdkit.Chem import Descriptors
print("El numero de aceptores de hidrgeno es: ",Descriptors.NumHAcceptors(ibu))
print("El numero de dadores de hidrgeno es: ",Descriptors.NumHDonors(ibu))
print("El coeficiente de reparto octanol-agua, LogP es: ", Descriptors.MolLogP(ibu))
print("El peso molecular es: ", Descriptors.MolWt(ibu))
```

Como era de esperar, el ibuprofeno cumple con las RO5 de Lipinski. 


### Similaridad Qumica 

La nocin de similaridad qumica (o similaridad molecular), segn el cual las molculas se agrupan segn sus propiedades fisico-qumicas, juega un papel importante en la prediccin de las propiedades de molculas y en particular en estudios de diseo de frmacos. 

Estos estudios se basan en la premisa de que compuestos similares tendrn propiedades similares. 

La idea es disear compuestos "similares" a los compuestos activos conocidos con la esperanza de encontrar estructuras moleculares alternativas que mantengan la actividad, pero al mismo tiempo mejorando, por ejempo su biodisponibilidad. 

En la prctica estos estudios de diseo de frmacos comienzan realizando una busqueda por similaridad en bases de datos de compuestos qumicos. 

El clculo de la similaridad de dos molculas cualesquiera se logra comparando sus **fingerprints moleculares**. Estas fingerprints (fps) se componen de informacin estructural sobre la molcula que ha sido codificada como una serie de bits.

El clculo de la similaridad se realiza en **2 etapas**, primero se calculan las fps de las molculas en cuestion y luego se comparan las fps con un **coeficiente de similaridad**, para finalmente producir la medida de similaridad. 

Existen diferentes tipos de fps: basadas en subestructura (PubChemFP, MACCS, etc) y en topologa (APFP, GraphFP, Morgan, ECFP, etc). Aqu presentamos dos fps de uso comn: **MACCS keys** y **fingerprints de Morgan** 

### MACCS keys

Las MACCS (del ingles Molecular ACCess System), tambin denominadas claves estructurales MACCS (keys), constan de 166 fragmentos estructurales predefinidos. Cada posicin consulta la presencia o ausencia de un fragmento o "key" estructural en particular.


### Fingerprints de Morgan 

Esta familia de fps estn basadas en el algoritmo de Morgan. Los "bits" corresponden a los entornos circulares de cada tomo en una molcula. El nmero de atomos y enlaces circundantes se define por el radio. Otras fps muy populares que son derivadas usando una variante del algoritmo de Morgan son las ECFP (por Extended connectivity fingerprints). 




### Medida de similaridad molecular 

Una vez calculadas las fps, la similaridad molecular puede cuantificarse con una serie de coeficientes de similaridad, el mas conocido es el coefiente de Tanimoto: 

#### Coeficiente de Tanimoto 

$$T _{c}(A,B) = \frac{c}{a+b-c}$$

- a: nmero de caracteristicas presentes en la molcula A 
- b: nmero de caracteristicas presentes en la molcula B 
- c: nmero de caracteristicas compartidas por A y B


Veamos como calcular las fps y realizar las medidas de similaridad en **RDKit**.
Primero definamos y dibujemos 8 molculas de ejemplo. 

```python
# Molecules in SMILES format
molecule_smiles = [
    "CC1C2C(C3C(C(=O)C(=C(C3(C(=O)C2=C(C4=C1C=CC=C4O)O)O)O)C(=O)N)N(C)C)O",
    "CC1(C(N2C(S1)C(C2=O)NC(=O)C(C3=CC=C(C=C3)O)N)C(=O)O)C",
    "C1=COC(=C1)CNC2=CC(=C(C=C2C(=O)O)S(=O)(=O)N)Cl",
    "CCCCCCCCCCCC(=O)OCCOC(=O)CCCCCCCCCCC",
    "C1NC2=CC(=C(C=C2S(=O)(=O)N1)S(=O)(=O)N)Cl",
    "CC1=C(C(CCC1)(C)C)C=CC(=CC=CC(=CC(=O)O)C)C",
    "CC1(C2CC3C(C(=O)C(=C(C3(C(=O)C2=C(C4=C1C=CC=C4O)O)O)O)C(=O)N)N(C)C)O",
    "CC1C(CC(=O)C2=C1C=CC=C2O)C(=O)O",
]

# List of molecule names
molecule_names = [
    "Doxycycline",
    "Amoxicilline",
    "Furosemide",
    "Glycol dilaurate",
    "Hydrochlorothiazide",
    "Isotretinoin",
    "Tetracycline",
    "Hemi-cycline D",
]
```

Convertimos los SMILES en objetos Molecula de rdkit y luego las visualizamos con el mdulo Draw en rdkit. 
Ntese que en este caso, como tenemos una lista de 8 SMILES, es conveniente utilizar un **for loop** para convertir todas ellas de una sola vez.

```python
from rdkit import Chem 
ms = [Chem.MolFromSmiles(x) for x in molecule_smiles]
```

Para graficar las 8 molculas en una grilla usamos la funcion MolsToGridImage del modulo Draw.Ademas especificamos la los nombres de las molculas proporcionados en la lista molecule_names:

```python
from rdkit.Chem import Draw

Draw.MolsToGridImage(
    ms,
    molsPerRow=3,
    subImgSize=(450, 150),
    legends=molecule_names,
)
```

Ahora calculemos la similaridad entre estas molculas. Tomemos de ejemplo el antibitico Doxiciclina y calculemos sus MACCS fingerprints.  

```python
molecule1 = ms[0]
molecule1
```

```python
from rdkit.Chem import MACCSkeys
maccs_fp1 = MACCSkeys.GenMACCSKeys(molecule1)
```

La correspodiente cadena de bits de la fingerprint asociada a esta molcula ser: 

```python
maccs_fp1.ToBitString()
```

Ahora calculemos la fp de otra molcula, por ejemplo Amoxicilina y a continuacin calculemos la similaridad entre ambas, con el coeficiente de Tanimoto:

```python
molecule2 = ms[1]
maccs_fp2 = MACCSkeys.GenMACCSKeys(molecule2)

from rdkit import DataStructs
DataStructs.FingerprintSimilarity(maccs_fp1,maccs_fp2)
```

El indice de tanimoto va de 0 para molculas sin ninguna similaridad a 1 para molculas identicas. 

Ahora calculemos las fps de todas las molculas:

```python
maccs_fps = [MACCSkeys.GenMACCSKeys(x) for x in ms]
```

A continuacin, queremos comparar una molcula de consulta (query) con nuestra lista de molculas. Para ello usamos la funcin BulkTanimotoSimilarity: 

```python
# Define molecule query and list
molecule_query = maccs_fp1                # dociciclina 
molecule_list = maccs_fps                 # todas 

# Calculate similarty values between query and list elements
molecule_sim = DataStructs.BulkTanimotoSimilarity(molecule_query, molecule_list)
```

Veamos esta informacin en un Data frame de Pandas. Agreguemos dos columnas, una para el nombre de la molcula y otra pas sus correspodientes valores de similaridad con respecto a la molcula de consulta. Adems, para que sea mas fcil la interpretacin de los datos, ordenemos los filas de la tabla de mayor a menor indice de similaridad:  

```python
import pandas as pd
table = pd.DataFrame()
table['name'] = molecule_names
table['similarity'] = molecule_sim
table.sort_values('similarity', ascending=False)
```

La Tetraciclina es la molcula mas similar a nuestra molecula de consulta Dociciclina, lo cual tiene sentido porque ambos pertenecen a la misma familia de antibioticos.
